```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from astropy.io import fits
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans, DBSCAN
from sklearn_extra.cluster import KMedoids
from scipy.stats import gaussian_kde
import seaborn as sns

sns.set(style="darkgrid")
```

```python
df = fits.open('../Datafiles/mysterious-peaks.fits')

df.info()

points = df[0].data
points = points.reshape(-1, 1)
```

```python
A = (points.T[0] - min(points.T[0]))/(max(points.T[0]) - min(points.T[0]))
B = np.linspace(0, 1, len(points.T[0]))

X = np.array((A, B))
X = pd.DataFrame(X.T)
```

```python
plt.scatter(X[0], X[1], s=1)
plt.show()
```

```python
plt.figure(figsize=(4*1.2, 4))
counts, bins, _  = plt.hist(X[0], bins=50, density=True)
plt.xlabel('Points')
plt.ylabel('Density')
plt.show()
```

```python
# Set up a range of n_components to test
n_components_range = range(2, 15)

# Initialize a list to store the AIC and BIC scores for each model
aic_scores = []
bic_scores = []

#
X1D = X[0].values.reshape(-1, 1)

# Iterate over the range of n_components
for n_components in n_components_range:
    # Find the optimal number of clusters using K-means
    kmeans = KMeans(n_clusters=n_components, random_state=0, algorithm='elkan')
    kmeans.fit(X)
    n_components = len(kmeans.cluster_centers_)
    
    # Use the cluster centroids as initial estimates for the means
    means_init = kmeans.cluster_centers_
    
    # Fit the GMM with the current value of n_components
    gmm = GaussianMixture(n_components=n_components, 
                          means_init=means_init.T[0].reshape(n_components, 1), random_state=0, warm_start=True)
    gmm.fit(X1D)
    
    # Calculate the AIC score for the current model
    aic_scores.append(gmm.aic(X1D))
    
    # Calculate the AIC score for the current model
    bic_scores.append(gmm.bic(X1D))

# Find the index of the minimum AIC score
optimal_n_components_idx_AIC = np.argmin(aic_scores)
optimal_n_components_idx_BIC = np.argmin(bic_scores)
```

```python
plt.figure(figsize=(5*1.2, 5))
plt.plot(n_components_range, aic_scores, '--o', label='AIC')
plt.plot(n_components_range, bic_scores, '--o', label='BIC')
plt.scatter(n_components_range[optimal_n_components_idx_AIC], min(aic_scores), marker='X', c='r', s=200, zorder=99)
plt.scatter(n_components_range[optimal_n_components_idx_BIC], min(bic_scores), marker='X', c='r', s=200, zorder=99)
plt.xlabel('# components', fontsize=16)
plt.ylabel('Score', fontsize=16)
plt.legend(loc='best', fontsize=16)
plt.grid(True)
plt.show()
```

```python
# Retrieve the optimal number of components
optimal_n_components = 8
```

```python
kmeans = KMeans(n_clusters=optimal_n_components, algorithm='elkan')
kmeans.fit(X)
means_init = kmeans.cluster_centers_

plt.scatter(X[0], X[1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_.T[0], kmeans.cluster_centers_.T[1], c='w', marker='X', s=50)
plt.show()
```

```python
gmm = GaussianMixture(n_components=optimal_n_components, 
                      means_init=means_init.T[0].reshape(optimal_n_components, 1), random_state=0, warm_start=True)
gmm.fit(X1D)

# Generate samples from GMM
samples = gmm.sample(len(X1D)*2)[0]

# Plot results
plt.hist(X1D, bins=50, density=True, label='Histogram')
x_plot = np.linspace(min(X[1]), max(X[1]), 1000)
y_plot = np.exp(gmm.score_samples(x_plot.reshape(-1, 1)))
plt.plot(x_plot, y_plot, '-r', label='GMM fit')
#plt.hist(samples.T[0], bins=50, density=True, label='GMM samples')
plt.legend()
plt.show()
```

```python
plt.scatter(X[0], X[1], s=2, alpha=0.5)
plt.scatter(samples.T[0], np.linspace(min(X[1]), max(X[1]), len(samples.T[0])), s=2, alpha=0.5)
plt.show()
```

```python
# Set up a range of n_components to test
n_components_range = range(2, 15)

# Initialize a list to store the AIC and BIC scores for each model
aic_scores = []
bic_scores = []

#
X1D = X[0].values.reshape(-1, 1)

# Iterate over the range of n_components
for n_components in n_components_range:
    # Find the optimal number of clusters using K-means
    kmedoids = KMedoids(n_clusters=n_components, metric='l1', method='pam')
    kmedoids.fit(X)
    n_components = len(kmedoids.cluster_centers_)
    
    # Use the cluster centroids as initial estimates for the means
    means_init = kmedoids.cluster_centers_
    
    # Fit the GMM with the current value of n_components
    gmm = GaussianMixture(n_components=n_components, 
                          means_init=means_init.T[0].reshape(n_components, 1), random_state=0, warm_start=True)
    gmm.fit(X1D)
    
    # Calculate the AIC score for the current model
    aic_scores.append(gmm.aic(X1D))
    
    # Calculate the AIC score for the current model
    bic_scores.append(gmm.bic(X1D))

# Find the index of the minimum AIC score
optimal_n_components_idx_AIC = np.argmin(aic_scores)
optimal_n_components_idx_BIC = np.argmin(bic_scores)
```

```python
plt.figure(figsize=(5*1.2, 5))
plt.plot(n_components_range, aic_scores, '--o', label='AIC')
plt.plot(n_components_range, bic_scores, '--o', label='BIC')
plt.scatter(n_components_range[optimal_n_components_idx_AIC], min(aic_scores), marker='X', c='r', s=200, zorder=99)
plt.scatter(n_components_range[optimal_n_components_idx_BIC], min(bic_scores), marker='X', c='r', s=200, zorder=99)
plt.xlabel('# components', fontsize=16)
plt.ylabel('Score', fontsize=16)
plt.legend(loc='best', fontsize=16)
plt.grid(True)
plt.show()
```
