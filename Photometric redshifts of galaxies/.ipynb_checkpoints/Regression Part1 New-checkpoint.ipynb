```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from astropy.table import Table
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.svm import OneClassSVM
from sklearn.decomposition import PCA

sns.set(style="darkgrid")
```

```python
tA = pd.read_csv('../Datafiles/PhotoZFileA.csv')
tB = pd.read_csv('../Datafiles/PhotoZFileB.csv')

display(tA)
display(tB)
```

```python
X = tA[['mag_r', 'u-g', 'g-r', 'r-i', 'i-z']]
y = tA['z_spec']

X_val = tB[['mag_r', 'u-g', 'g-r', 'r-i', 'i-z']]
y_val = tB['z_spec']
```

```python
## Define the function that we want to optimize
def error(y_true, y_pred):
    return np.median(abs((y_true - y_pred)/(1 + y_true)))

def neg_error(y_true, y_pred):
    return -error(y_true, y_pred)

custom_scorer = make_scorer(neg_error)
```

```python
## Linear regression
model = LinearRegression(fit_intercept=True)
cv_scores = cross_val_score(model, X, y, cv=10, scoring=custom_scorer)

print("Mean Score: %.5f" % -cv_scores.mean())
print("Std Score: %.5f" % cv_scores.std())

model.fit(X, y)
b = model.intercept_
w = model.coef_

y_pred = model.predict(X).T
Error = error(y, y_pred)
print('\nError on tA: %.5f ' % Error)

y_pred = model.predict(X_val).T
Error = error(y_val, y_pred)
print('Error on tB: %.5f ' % Error)
```

```python
def corr_cornerplot(df, labels):
    plt.figure(figsize=(18, 18), layout='tight')
    cmap = plt.cm.cividis
    cmap.set_under(color='w')
    
    corr = df.corr()
    n = len(labels)
    for i in range(0, n):
        for j in range(0, n):
            if i >= j:
                plt.subplot(n, n, i*n+j+1)
                if i == j:
                    plt.hist(df[labels[i]], bins=100, density=True, histtype='step')
                    plt.xlabel(labels[i], fontsize=14)
                    if j == 0:
                        plt.ylabel('Density', fontsize=14)
                else:
                    plt.title('Corr: ' + repr(round(corr.loc[labels[j]][labels[i]], 3)), fontsize=12)
                    plt.hist2d(df[labels[j]], df[labels[i]], bins=80, cmap=cmap, vmin=1)
                    if j == 0:
                        plt.ylabel(labels[i], fontsize=14)
                    plt.xlabel(labels[j], fontsize=14)
    plt.show()
```

```python
corr_cornerplot(tA, tA.columns)
corr_cornerplot(tB, tB.columns)
```

```python
## Remove outliers
def remove_outliers(df, nu):
    clf = OneClassSVM(nu=nu)
    clf.fit(df)
    y_pred = clf.predict(df)
    outliers = df[y_pred == -1]

    df = df[~df.isin(outliers)].dropna()
    df = df.reset_index(drop=True)
    
    return df, outliers
```

```python
## Remove outliers from Table A
tA_cut, tA_cut_outliers = remove_outliers(tA, 0.5)

## Remove outliers from Table B
#tB_cut, tB_cut_outliers = remove_outliers(tB, 0.1)
```
